```python
import pandas as pd

telecom = pd.json_normalize(datos.to_dict(orient='records'))

telecom.head()
```

```python
telecom.columns
```

```python
telecom['account.Charges.Total'] = pd.to_numeric(
    telecom['account.Charges.Total'],
    errors='coerce'
)
```

```python
telecom['account.Charges.Total'].dtype
```

```python
telecom['account.Charges.Total'].isnull().sum()
```

```python
telecom['account.Charges.Total'] = telecom['account.Charges.Total'].fillna(0)
```

```python
telecom['account.Charges.Total'].isnull().sum()
```

```python
telecom['Cuentas_Diarias'] = telecom['account.Charges.Monthly'] / 30
```

```python
telecom[['account.Charges.Monthly', 'Cuentas_Diarias']].head()
```

```python
telecom.columns
```

```python
telecom['Cuentas_Diarias'].describe()
```

```python
binarios = {
    "Yes": 1,
    "No": 0
}

telecom["Churn"] = telecom["Churn"].map(binarios)
telecom["phone.PhoneService"] = telecom["phone.PhoneService"].map(binarios)
telecom["account.PaperlessBilling"] = telecom["account.PaperlessBilling"].map(binarios)
```

```python
telecom[["Churn","phone.PhoneService","account.PaperlessBilling"]].head()
```

```python
telecom["customer.SeniorCitizen"] = telecom["customer.SeniorCitizen"].astype(int)
```

```python
telecom = telecom.rename(columns={
    "customer.gender": "Genero",
    "customer.SeniorCitizen": "AdultoMayor",
    "customer.Partner": "TienePareja",
    "customer.Dependents": "Dependientes",
    "customer.tenure": "MesesCliente",
    "phone.PhoneService": "ServicioTelefono",
    "phone.MultipleLines": "MultiplesLineas",
    "internet.InternetService": "ServicioInternet",
    "account.Contract": "TipoContrato",
    "account.PaperlessBilling": "FacturacionDigital",
    "account.PaymentMethod": "MetodoPago",
    "account.Charges.Monthly": "CargoMensual",
    "account.Charges.Total": "CargoTotal"
})
```

#Informe Final Análisis de Evasión de Clientes (Churn)
##Introducción

El objetivo de este análisis fue investigar la evasión de clientes (Churn) en Telecom X, identificando patrones y factores asociados a la cancelación del servicio. Comprender estos factores permite apoyar estrategias de retención y el desarrollo de modelos predictivos para anticipar la pérdida de clientes.

##Limpieza y Tratamiento de Datos

Los datos fueron importados desde una API en formato JSON y convertidos en un DataFrame de Pandas. Posteriormente se normalizó la estructura anidada para obtener un dataset tabular.

Durante esta etapa se realizaron las siguientes acciones:

inspección de columnas y tipos de datos

detección y tratamiento de valores nulos

verificación y eliminación de registros duplicados

conversión de variables a formatos adecuados para análisis

creación de la variable Cuentas_Diarias a partir de la facturación mensual

Estas transformaciones permitieron preparar los datos para el análisis exploratorio.

##Análisis Exploratorio de Datos

Se realizaron estadísticas descriptivas y visualizaciones para comprender la distribución del churn y su relación con otras variables.

Se analizaron:

la proporción de clientes que cancelaron el servicio

la distribución del churn según variables categóricas como género, tipo de contrato y método de pago

la relación entre churn y variables numéricas como antigüedad del cliente, cargos mensuales y gasto total

Los gráficos permitieron identificar diferencias relevantes entre clientes que cancelan y aquellos que permanecen en el servicio.

##Conclusiones e Insights

El análisis reveló algunos factores asociados a la evasión de clientes:

clientes con contratos mensuales presentan mayor probabilidad de churn

clientes con menor antigüedad tienden a cancelar con mayor frecuencia

clientes con cargos mensuales más altos presentan mayor tendencia a cancelar

los clientes que cancelan suelen tener menor gasto total acumulado

Estos patrones sugieren que la experiencia inicial del cliente y el tipo de contrato influyen significativamente en la evasión.

##Recomendaciones

Con base en los resultados obtenidos, se sugieren las siguientes acciones:

reforzar estrategias de retención en los primeros meses del cliente

incentivar contratos de mayor duración

evaluar el impacto de los cargos mensuales en la satisfacción del cliente

utilizar estos hallazgos para desarrollar modelos predictivos de churn

Estas medidas pueden ayudar a Telecom X a reducir la evasión y mejorar la fidelización de sus clientes.
